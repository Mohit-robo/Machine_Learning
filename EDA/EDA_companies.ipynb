```python
df_companies.drop(df_companies.index[df_companies.name.isna()], inplace=True)
df_companies.drop(df_companies.index[df_companies.industry.isna()], inplace=True)
```

```python
df_companies.drop(df_companies.index[df_companies.domain.isna()], inplace=True)
df_companies.drop(df_companies.index[df_companies.year_founded.isna()], inplace=True)
df_companies.drop(df_companies.index[df_companies.locality .isna()], inplace=True)
df_companies.drop(df_companies.index[df_companies.country.isna()], inplace=True)
```

```python
df_companies.rename(columns={'year founded':'year_founded', 'linkedin url':'linkedin_url',
                             'current employee estimate':'current_employee_estimate',
                             'total employee estimate':'total_employee_estimate'},inplace= True)
```

```python
sns_plot_year = sns.distplot(df_companies["year_founded"])
sns_plot_year.set_xlim(1900,2020)  # -- > left skewed
```

```python
df_corr = df_companies.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(df_corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f_fig, f_ax = plt.subplots(figsize=(12, 12))
    f_ax = sns.heatmap(df_corr, mask=mask, vmin=0, vmax=1, square=True,
                        annot=True, annot_kws={"size": 10}, cmap="BuPu")
plt.title('Correlation of numeric features')
plt.show()
```
